```python
# Import the dependencies.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
lats = np.random.uniform(low=-90.000, high=90.000, size=2000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2000)
```

```python
coords = zip(lats,lngs)
coordinates = list(coords)
```

```python
from citipy import citipy
```

```python
cities=[]

for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0],coordinate[1]).city_name
    if city not in cities:
        cities.append(city)
```

```python
len(cities)
```

```python
from config import api_key

```

```python
link = "http://api.openweathermap.org/data/2.5/weather?units=Imperial"

url = link + "&appid=" + api_key 
```

```python
import requests
```

```python
city_data= []


for i in cities:
    city_url = url + "&q=" + i.replace(" ","+")
    try:
        city_weather = requests.get(city_url).json()
        country = city_weather["sys"]["country"]
        lat = city_weather["coord"]["lat"]
        lon = city_weather["coord"]["lon"]
        max_temp = city_weather["main"]["temp_max"]
        humidity = city_weather["main"]["humidity"]
        cloudiness = city_weather["clouds"]["all"]
        wind_Speed = city_weather["wind"]["speed"]
        weather_desc = city_weather["weather"][0]["description"]
        
        city_data.append({"City": i.title(),
                          "Country": country,
                          "Lat": lat,
                          "Lon": lon,
                          "Max Temp": max_temp,
                          "Humidity" : humidity,
                          "Cloudiness" : cloudiness,
                          "Wind Speed" : wind_Speed,
                          "Weather description" : weather_desc})
    except Exception as error:
        print(error)
```

```python
city_data_df = pd.DataFrame(city_data)
```
